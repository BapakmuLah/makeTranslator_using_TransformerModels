```python
# IMPORT LIBRARY FOR LOAD MODEL

from transformers import pipeline
```

LINK TRANSFORMERS MODEL :
https://huggingface.co/facebook/nllb-200-distilled-600M

```python
# LOAD A MODEL

translator = pipeline(task='translation', model='facebook/nllb-200-distilled-600M')
translator
```

```python
# CREATE PROMPT

prompt_1='i lost everything. im found in unconsiousness'

prompt_2='''The ineffable intricacies of human cognition, coupled with the paradoxical interplay between conscious deliberation and
unconscious impulse, present an enigmatic challenge for contemporary philosophers and neuroscientists alike, as they endeavor to decipher
the elusive mechanisms governing thought, emotion, and behavior, all while grappling with the inherent limitations of language itself
to accurately encapsulate the multifaceted nature of the mind in its totality."This sentence is dense with abstract concepts,
making it difficult to understand without careful thought and background knowledge in philosophy and neuroscience'''

prompt_3='''Amidst the ceaseless oscillations of societal norms, where fleeting trends and entrenched ideologies collide in an eternal dance of contradiction,
one must ponder whether true autonomy is an attainable ideal or merely an illusion crafted by the very systems that purport to emancipate the individual,
all while acknowledging the profound limitations of human perception that obscure the true nature of freedom and self-determination.
This sentence presents a challenging mix of philosophical ideas and abstract reasoning'''
```

```python
# CONVERT FROM ENGLISH TO INDONESIA LANGUAGE

indo_translated_1 = translator(prompt_1, src_lang='eng_Latn', tgt_lang='ind_Latn')

indo_translated_2 = translator(prompt_2, src_lang='eng_Latn', tgt_lang='ind_Latn')

indo_translated_3 = translator(prompt_3, src_lang='eng_Latn', tgt_lang='ind_Latn')
```

```python
indo_translated_2
```

```python
indo_translated_3
```

```python
# CONVERT FROM INDO TO ENGLISH LANGUAGE

prompt_4 = 'Aku kehilangan segalanya. Aku ditemukan dalam kesesatan.'

prompt_5 = '''Kerumitan tak terungkap dari kognisi manusia, ditambah dengan interaksi paradoksal antara pertimbangan sadar dan dorongan tak sadar, menyajikan tantangan
yang misterius bagi filsuf kontemporer dan ahli saraf sama-sama, karena mereka berusaha untuk memecahkan mekanisme yang sulit dipahami yang mengatur pemikiran, emosi, dan perilaku,
sementara berjuang dengan keterbatasan inheren bahasa itu sendiri untuk secara akurat mengkapsulasi sifat bermacam-macampikiran dalam totalitasnya".
Undang-undang ini padat dengan konsep abstrak, sehingga sulit dipahami tanpa pemikiran yang cermat dan pengetahuan latar belakang dalam filsafat dan ilmu saraf'''

prompt_6 = ''' Di tengah bergeser-geser tanpa henti dari norma-norma masyarakat, di mana tren-tren yang tak terpisahkan dan ideologi-ideologi yang berakar berantakan
dalam tarian abadi kontradiksi, seseorang harus merenungkan apakah otonomi sejati adalah ide yang dapat dicapai atau hanya ilusi yang dibuat oleh sistem yang bertujuan
untuk membebaskan individu, semua sambil mengakui keterbatasan mendalam persepsi manusia yang membungkam sifat sejati kebebasan dan penentuan nasib sendiri. '''
```

```python

eng_translated_1 = translator(prompt_4, src_lang='ind_Latn', tgt_lang='eng_Latn')

eng_translated_2 = translator(prompt_5, src_lang='ind_Latn', tgt_lang='eng_Latn')

eng_translated_3 = translator(prompt_6, src_lang='ind_Latn', tgt_lang='eng_Latn')
```
